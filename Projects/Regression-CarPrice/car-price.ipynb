```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("CarPrice_Assignment.csv")
df.head()
```

```python
df.info()
```

```python
df.drop("car_ID",axis=1)
```

```python
df["CarName"]=df["CarName"].apply(lambda x: x.split()[0]  )

df
```

```python
df.columns
```

```python
X = df.drop("price",axis=1)
y = df["price"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer 

categorical_features = ['CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem']

one_hot=OneHotEncoder()

transformer = ColumnTransformer( [ ("one_hot",one_hot,categorical_features)],
                               remainder="passthrough")

transformed_x = transformer.fit_transform(X)


```

```python
# Obtener el OneHotEncoder entrenado dentro del ColumnTransformer
one_hot_fitted = transformer.named_transformers_["one_hot"]

# convertir nuevamente a dataframe
column_names = one_hot_fitted.get_feature_names_out(categorical_features)
column_names.tolist()
```

```python
new_X = pd.DataFrame(transformed_x, columns=column_names.tolist() + X.drop(columns=categorical_features).columns.tolist())


new_X.head() # Ver resultado
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size=0.2)

np.random.seed(42)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Diccionario con modelos
modelos = {
    "Regresión Lineal": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.1),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
}

# Entrenar y evaluar cada modelo
resultados = {}
for nombre, modelo in modelos.items():
    print(f"Entrenando {nombre}...")
    
    # Entrenar modelo
    modelo.fit(X_train_scaled, y_train)
    
    # Hacer predicciones
    y_pred = modelo.predict(X_test_scaled)
    
    # Calcular métricas
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    # Guardar resultados
    resultados[nombre] = {
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R²": r2
    }

# Mostrar los resultados
for modelo, metricas in resultados.items():
    print(f"\n🔹 {modelo}:")
    for metrica, valor in metricas.items():
        print(f"   {metrica}: {valor:.4f}")
```
