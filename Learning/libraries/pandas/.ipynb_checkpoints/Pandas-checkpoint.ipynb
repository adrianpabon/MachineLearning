```python
missing_data = os.path.abspath('../../../files/car-sales-missing-data.csv')
missing_data = pd.read_csv(missing_data)
missing_data
```

```python
missing_data["Odometer"]=missing_data["Odometer"].fillna(missing_data["Odometer"].mean())
missing_data
```

```python
missing_data.dropna(inplace=True)
```

```python
missing_data.to_csv("cars_clean.csv")

car_clean = pd.read_csv("cars_clean.csv")
car_clean
```

```python
### making a column from a series

seats_column = pd.Series([5,5,5,5,5])
car_clean["Seats"]=seats_column
car_clean
```

```python
car_clean["Seats"].fillna(5)
```

```python
## column from Python list
import numpy as np

fuel_economy = np.random.uniform(4,7,6)
car_clean["fuel"]=fuel_economy
car_clean
```

```python
car_clean["Total fuel"] = car_clean["Odometer"]*car_clean["fuel"]
car_clean
```

```python
# Create a column from a single value
car_clean["Number of wheels"] = 4
car_clean
```

```python
car_clean = car_clean.drop("Number of wheels", axis=1)
```

```python
car_clean_shuffled = car_clean.sample(frac=0.2) #here we are taking the 20% of the df samples
car_clean_shuffled
```

```python
car_clean["Odometer"] = car_clean["Odometer"].apply(lambda x: x/1.6)
car_clean
```
